```python
js_text_template = Template('''
var margin = {top:20,right:20,bottom:30,left:30};
var width = 850 - margin.left - margin.right;
var height = 300 - margin.top - margin.bottom;


var x = d3.scale.linear().range([0,width]);


var y = d3.scale.linear().range([height,0]);

var color = d3.scale.category10();

var xAxis = d3.svg.axis()
.scale(x)
.orient("bottom");

var yAxis = d3.svg.axis()
.scale(y)
.orient("left");

var svg = d3.select("#$maid").append("svg")
.attr("width", width + margin.left+margin.left)
.attr("height", height + margin.top+margin.bottom)
.append("g")
.attr("transform","translate("+margin.left+","+margin.top+")")

var data = $python_data;

data.forEach(function(d) {

});



x.domain(d3.extent(data,function(d) { return d.Currency;})).nice();
y.domain(d3.extent(data,function(d) { return d.KZT;})).nice();



console.log("Extent",d3.extent(data,function(d) { return d.Currency}));

var barWidth = (width / data.length);
var barPadding = 5;

svg.append("g")
.attr("class","x axis")
.attr("transform","translate(40,"+height+")")
.call(xAxis)
.append("text")
.attr("class","label")
.attr("x",width)
.style("text-anchor","end")
.text("Currency");

svg.append("g")
.attr("class","y axis")
.call(yAxis)
.append("text")
.attr("transform","rotate(-90)")
.attr("class","label")
.attr("y",6)
.attr("dy",".8em")
.style("text-anchor","end")
.text("KZT, hundred");

svg.selectAll(".bar")
.data(data)
.enter()
.append("rect")
.classed('bar',true)
.attr('width',barWidth - barPadding)
.attr('height',function(d) { return height - y(d.KZT);})
.attr("x", function(d) { return x(d.Currency);})
.attr("y", function(d) { return y(d.KZT);})
.attr("transform", function (d, i) {
        var translate = [barWidth * i, 0];
        return "translate("+ translate +")";
    });
    
''');


```

```python
html_template = Template('''
    <style>
    .rect{
        stroke: #000;
    }
    .legend{
        color:blue;
    }
    
    .axis path,
    .axis line{
        fill: none;
        shape-rendering:crispEdges;
    
    }
    </style>
    <div id="scatter-id"></div>
    <script>$js_text</script>
''')
js_text = js_text_template.substitute({
    "python_data":json.dumps(df_dict),
    "maid":"scatter-id"
})

HTML(html_template.substitute({'js_text':js_text}))
```
